```python
import numpy as np
import pandas as pd
```

```python
src_path_log = '../data/log/robotdata1.log'
logfile = open(src_path_log, 'r')
all_readings = []
for time_idx, line in enumerate(logfile):
    meas_type = line[0]
    if (meas_type == "L"):
        meas_vals = np.fromstring(line[2:], dtype=np.float64, sep=' ')
        odometry_laser = meas_vals[3:6]
        ranges = meas_vals[6:-1]
        all_readings.append(ranges)

all_readings = pd.Series(np.vstack(all_readings).flatten())
    
```

```python
bins = list(np.arange(0,1040,50))
bins.append(3000)
all_readings.plot.hist(bins = bins)
```
